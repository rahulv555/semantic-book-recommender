# PLOTTING MISSING DATA
Missing values could be related to some bias. i.e. missing values are linked to specific charectiristc of the row

```python
axes = plt.axes() #instantiating matplotlib plot
sns.heatmap(books.isna().transpose(), cbar=False, ax=axes)

plt.xlabel("Columns")
plt.ylabel("Missing values")
```

we can see that for (average_rating, num_pages, ratings_count), if one is missing for a row, all are missing
This could be bias,
for eg: it could be that these rows were fetched from a another books dataset, and they didn't have these columns
for eg: could be newer books

```python
#1 if book is missing description, 1 in column(description is what we are going to use mainly for our NLP stuff)
books["missing_description"] = np.where(books["description"].isna(), 1, 0)
books["age_of_book"] = 2025 - books["published_year"]
```

```python
columns_of_interest= ["num_pages", "average_rating", "missing_description", "age_of_book"]

#seeing if any correlation between these
correlation_matrix = books[columns_of_interest].corr(method="spearman")  #Pearson more suitable for complete continuous variables, but description here is binary
sns.set_theme(style="white")
plt.figure(figsize=(8, 6))
heatmap  = sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", cbar_kws={"label": "Spearman Correlation"})
heatmap.set_title("Correlation Heatmap")
plt.show()
```

### We can see not much of correlation of books with missing description, missing the other 3 columns, with any of these other factors. HENCE, not much of an issue/bias

So we just delete missing data books to reduce headache

```python
#seeing number of rows we would be deleting
```

```python
books[(books["description"].isna()) | (books["num_pages"].isna()) | (books["average_rating"].isna()) | (books["published_year"].isna())
]
```

```python
book_missing = books[~(books["description"].isna()) & ~(books["num_pages"].isna()) & ~(books["average_rating"].isna()) & ~(books["published_year"].isna())
]
```

## Also, we can see lots of different categories, which is not ideal for a recommender. Would be good to delete some unwanted/too specific ones

```python
book_missing["categories"].value_counts().reset_index().sort_values("count", ascending=False)
```

We need to classify the not so frequent categories into the more common categories, so that the unique count is reduced. We can do this by text classification

# REMOVING SHORT DESCRIPTIONS

description without much info, such as one word, are not gonna be use full.

```python
book_missing
```

```python
book_missing = book_missing.copy()
book_missing["words_in_description"] = book_missing["description"].str.split().str.len()
```

```python
book_missing
```

No apparent cutoff, so we have to put arbitarily

```python
book_missing.loc[book_missing["words_in_description"].between(1, 4), "description"]
```

```python
book_missing.loc[book_missing["words_in_description"].between(5, 14), "description"]
```

```python
book_missing.loc[book_missing["words_in_description"].between(15, 24), "description"]
```

```python
book_missing.loc[book_missing["words_in_description"].between(25, 34), "description"]
```

We use 25 as cutoff, as it seems enough

```python
book_missing_25_words = book_missing[book_missing["words_in_description"]>=25]
```

```python
book_missing_25_words = book_missing_25_words.copy()
```

### Now, subtitles has a lot of missing value
so, we make new column to combine them


```python
book_missing_25_words["title_and_subtitle"] = (np.where(book_missing_25_words["subtitle"].isna(), book_missing_25_words["title"], book_missing_25_words[["title", "subtitle"]].astype(str).agg(": ".join, axis=1)))
book_missing_25_words
```

```python
book_missing_25_words["tagged_description"] = book_missing_25_words[["isbn13", "description"]].astype(str).agg(" ".join, axis=1)
```

```python
book_missing_25_words.drop(["subtitle", "missing_description", "age_of_book","words_in_description"], axis=1).to_csv("books_cleaned.csv", index=False)
```
